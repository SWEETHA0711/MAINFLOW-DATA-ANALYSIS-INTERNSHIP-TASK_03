```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the dataset
data=pd.read_csv("/content/householdtask3.csv")
data
```

```python
#displaying first 10 rows
data.head(10)
```

```python
print(f'no of rows:',data.shape[0])
print(f'no of columns :',data.shape[1])
```

```python
#checking for null values
data.isnull().sum()
```

```python
#checking for duplicate rows
data.drop_duplicates()
```

```python
#visualize data using scatterplot
plt.scatter(data['year'],data['own'],color='purple')

#title of the plot
plt.title('scatterplot')

#label for the exes
plt.xlabel('year')
plt.ylabel('own')

#displaying the plot
plt.show()
```

```python
#visualize data using lineplot
plt.plot(data['year'],color='orangered')
plt.plot(data['own'],color='green')

#setting legend
plt.legend(['year','own'])

#title of the plot
plt.title('lineplot')

#label for the axes
plt.xlabel('year')
plt.ylabel('own')

#displaying the plot
plt.show()
```

```python
#visualize data using barchart
plt.bar(data['year'],data['own'],color='cyan')

#title of the chart
plt.title('bar chart')

#label for the axes
plt.xlabel('year')
plt.ylabel('own')

#displaying the plot
plt.show()
```

```python
#visualize data using histogram
plt.hist(data['income'],color='magenta')

#title of the chart
plt.title('histogram')

#displaying the chart
plt.show()
```
